```python
from glob import glob

carla_v5i_img_list = glob('/content/drive/MyDrive/ColabProReposit/yolo_for_project/yolo8_dataset/carla_dataset.v5i.yolov8/train/images/*.jpg')
carla_v5i_label_list = glob('/content/drive/MyDrive/ColabProReposit/yolo_for_project/yolo8_dataset/carla_dataset.v5i.yolov8/train/labels/*.txt')

len(carla_v5i_img_list), len(carla_v5i_label_list)
```

```python
from sklearn.model_selection import train_test_split

train_img_list, val_img_list = train_test_split(carla_v5i_img_list, test_size=0.2, random_state=13)
len(train_img_list), len(val_img_list)
```

```python
train_txt = '/content/drive/MyDrive/ColabProReposit/yolo_for_project/yolo8_dataset/carla_dataset.v5i.yolov8/'
val_txt = '/content/drive/MyDrive/ColabProReposit/yolo_for_project/yolo8_dataset/carla_dataset.v5i.yolov8/'

with open(train_txt + 'train.txt','w') as f:
  f.write('\n'.join(train_img_list))

with open(val_txt + 'val.txt','w') as f:
  f.write('\n'.join(val_img_list))
```

```python
import yaml

# yaml ÌååÏùºÏóê Ï†ÄÏû•Ìï† Ï†ïÎ≥¥Î•º ÎîïÏÖîÎÑàÎ¶¨Î°ú Ï†ïÏùò

data={
    'train': train_txt + 'train.txt',
    'val': val_txt + 'val.txt',
    'nc': 7,
    'names': ['bike', 'green', 'pedesatrian', 'red', 'speed_limit', 'vehicle', 'yellow']
}

# ÎîïÏÖîÎÑàÎ¶¨ Ï†ïÎ≥¥Î•º yaml ÌååÏùºÏóê Ï†ÄÏû•

with open(train_txt + 'carla_v5i_data.yaml','w') as f:
  yaml.dump(data, f)
```
