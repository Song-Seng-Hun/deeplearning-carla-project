```python
from glob import glob

carla_img_list = glob('/home/syh/yolo/test3_plus/train/images/*.jpg')
carla_label_list = glob('/home/syh/yolo/test3_plus/train/labels/*.txt')

len(carla_img_list), len(carla_label_list)
```

```python
from sklearn.model_selection import train_test_split

train_img_list, val_img_list = train_test_split(carla_img_list, test_size=0.2, random_state=13)

len(train_img_list), len(val_img_list)
```

```python
with open('/home/syh/yolo/test3_plus/train/train.txt', 'w') as f :
  f.write('\n'.join(train_img_list) + '\n')

!cat /home/syh/yolo/test3_plus/train/train.txt | head -n 5
```

```python
with open('/home/syh/yolo/test3_plus/train/val.txt', 'w') as f :
  f.write('\n'.join(val_img_list) + '\n')

!cat /home/syh/yolo/test3_plus/train/val.txt | head -n 5
```
