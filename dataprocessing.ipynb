## change label

```python
test_label_map = {'1':'0', '2':'3'}
test_delete_map = {'0', '3'}

def change_label(line, label_map, delete_map):
    parts = line.strip().split()
    if len(parts) < 1:
        return None  # skip empty lines
    first_num = parts[0]
    if first_num in delete_map:
        return None  # delete the line
    new_num = label_map.get(first_num)
    if new_num is None:
        return line  # no change
    parts[0] = new_num
    return ' '.join(parts) + '\n'
print(change_label('0 123 123 123 123', test_label_map, test_delete_map))
print(change_label('2 123 123 123 123', test_label_map, test_delete_map))
```

```python
txt_files1 =  glob(path1 + '*.txt')
# iterate through each txt file and apply the label conversion function
for txt_file in tqdm(txt_files1):
    with open(txt_file, 'r+') as f:
        lines = f.readlines()
        f.seek(0) # move the cursor to the beginning of the file
        f.truncate() # truncate the file to zero length
        for line in tqdm(lines):
            new_line = change_label(line, label_map1, delete_map1)
            if new_line: # check if the new line is not None
                f.write(new_line)
```

```python
# 라벨에 사람이 있으면 다른 경로로 이미지 라벨 복사
train_data_num = 0

for file in os.listdir(label_dir1_train):
    if file.endswith('.txt') :
        txt_file_path = os.path.join(label_dir1_train, file)
        jpg_file_path = os.path.join(images_dir1_train, file.replace('.txt', '.jpg'))

        print(txt_file_path)
        # # .txt 파일 내용 중 첫 번째 줄이 1인지 확인
        with open(txt_file_path, 'r') as f:
            lines = f.readlines()
            for line in lines :
                if line == None or len(line)<= 0 :
                    break
                if line[0] == '1' :
                    shutil.copy(txt_file_path, target_dir_train + "/labels")
                    shutil.copy(jpg_file_path, target_dir_train + "/images")
                    train_data_num += 1
                    break
```
